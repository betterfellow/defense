```python
# Oct05_17-23-19_skywalkerui-MacBook-Pro.local
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torch.autograd import Variable
from tensorboard import SummaryWriter
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
% matplotlib inline
```

## Set Hyperparameter

```python
batch_size=64
learning_rate=0.002
num_epoch=15
leak=0.05
drop_rate=0.2
z_dim=20
```

## Data: mnist

```python
# Download data
mnist_train = dset.MNIST("./", train=True, transform=transforms.ToTensor(), target_transform=None, download=True)
mnist_test = dset.MNIST("./", train=False, transform=transforms.ToTensor(), target_transform=None, download=True)
```

```python
train_loader=torch.utils.data.DataLoader(mnist_train,batch_size=batch_size, shuffle=True, num_workers=2,drop_last=True)
test_loader=torch.utils.data.DataLoader(mnist_test,batch_size=batch_size,shuffle=False,num_workers=2,drop_last=True)
```

## Model
* The whole model is composed by 2 parts: Encoder & Decoder
* Xavier/He initialization
* Batch Normalization
* Dropout
* Encoder & Decoder
* rather than BCELoss() use BCEWithLogitsLoss()
    * More numerically stable [[link]](http://pytorch.org/docs/master/nn.html#torch.nn.BCEWithLogitsLoss)

### Encoder
* Two outputs
    * z_mu
    * z_stddev

```python
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder,self).__init__()
        
        self.fc_layer=nn.Sequential(
            nn.Linear(28*28,512),
            nn.BatchNorm1d(512),
            nn.LeakyReLU(leak),
            nn.Dropout(p=drop_rate,inplace=True),
            
            nn.Linear(512,256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(leak),
            nn.Dropout(p=drop_rate,inplace=True),
            
            nn.Linear(256,128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(leak),
        )
        
        
        # ===== Two output vectors generated by the Encoder =====
        
        # One for z_mu
        self.z_mu=nn.Sequential(
            nn.Linear(128,z_dim),
            nn.LeakyReLU(leak)
        )
        
        # Another for z_stddev
        self.z_stddev=nn.Sequential(
            nn.Linear(128,z_dim),
            nn.LeakyReLU(leak)
        )
        
        
    def forward(self,x):
        x = x.view(batch_size,-1)
        out=self.fc_layer(x)
        z_mu=self.z_mu(out)
        z_stddev=self.z_stddev(out)
        
        return z_mu,z_stddev
```

### Decoder
* input: sampled_z
* output: reconstructed image

```python
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder,self).__init__()
        
        self.fc_layer=nn.Sequential(
            nn.Linear(z_dim,128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(leak),
            nn.Dropout(p=drop_rate,inplace=True),
            
            nn.Linear(128,256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(leak),
            nn.Dropout(p=drop_rate,inplace=True),
            
            nn.Linear(256,512),
            nn.BatchNorm1d(512),
            nn.LeakyReLU(leak),
            nn.Dropout(p=drop_rate,inplace=True),
            
            nn.Linear(512,28*28),
            nn.Sigmoid()
        )
    
    # decode the sampled_z
    def forward(self,sampled_z):
        out=self.fc_layer(sampled_z)
        
        return out
```

### the Model as a whole
* encoder & decoder
* sample z

```python
class VAE(nn.Module):
    def __init__(self, encoder, decoder):
        super(VAE,self).__init__()
        
        self.encoder=encoder
        self.decoder=decoder
    
    def sample_z(self, z_mu,z_stddev):
        
        # sample epsilon ~ N(0, 1)
        epsilon=Variable(torch.randn(batch_size,z_dim),requires_grad=False)
        
        # reparameterization trick
        sampled_z=z_mu+torch.mul(torch.exp(z_stddev),epsilon)
       
        return sampled_z
        
            
    def forward(self,x):
        # encode the latent variable from the image
        # sample z from the encoded result
        # decode the sampled z
        
        z_mu,z_stddev=self.encoder(x)
        
        self.z_mu=z_mu
        self.z_stddev=z_stddev
        
        sampled_z=self.sample_z(z_mu,z_stddev)
        result=self.decoder(sampled_z)
        
        return result
```

```python
model=VAE(Encoder(),Decoder())
```

## Loss function
* Reconstruction error = mean(square(generated_image - real_image))  
* latent loss (or regularizer) = KL-Divergence(latent_variable, unit_gaussian)  
* Loss = latent_loss + reconstruction_error

[[kvfrans blog]](http://kvfrans.com/variational-autoencoders-explained/)

### Reconstruction loss (or error)

* Binary cross entropy
* In the case of Bernoulli distribution p(x|z), log-likelihood is cross entropy? [[link]](https://www.slideshare.net/HyungjooCho2/deep-generative-modelpdf)

\begin{equation*}
\mathbb{E}_{q_\phi(z|x)}log{p(z|x)} = \sum_i^n \left(x_i \ln p(y_i) + (1-x_i) \ln (1-p(y_i)\right)
\end{equation*}

### The latent loss (or regularizer)
* from the original paper: Appendix B

\begin{equation*}
-D_{KL}\left(q_\phi(z)\|p_\theta(z)\right) =\frac{1}{2}\sum_j \left( 1+log((\sigma_j )^2 )-(\mu_j)^2-(\sigma_j)^2 \right)
\end{equation*}


우변을 각각 mu와 stddev에 대해서 미분하면 각각 0과 1이 될 때 latent loss인 KL_loss가 최소가 된다.

```python
def loss_func(generated_image,x,z_mu,z_logvar):
    
    recon=nn.BCELoss(size_average=False)
    reconstruction_loss=recon(generated_image,x)
    
    # alias the KL-divergence term
    latent_loss=-0.5*torch.sum(1+z_logvar-z_mu**2-z_logvar.exp())
    
    return reconstruction_loss+latent_loss
```

## Optimizer

```python
optimizer=optim.Adam(model.parameters(),lr=learning_rate)
```

## Training

```python
writer=SummaryWriter()

for i in range(num_epoch):
    for batch_idx,[image,_] in enumerate(train_loader):
        n_iter = (i*len(train_loader))+batch_idx
        
        x=Variable(image)
        
        optimizer.zero_grad()
        
        reconstructed_image=model(x)
        
        loss=loss_func(reconstructed_image,x,model.z_mu,model.z_stddev)
        loss.backward()
        
        optimizer.step()
        
        if batch_idx%1000==0:
            print('Training Epoch: {} \tn_iter: {} \tLoss: {:.6f}'.format(
                i, n_iter, loss.data[0]))
            
        writer.add_scalar('loss',loss.data[0] / len(x),n_iter)
        
plt.imshow(model(x).data[0].numpy().reshape(28, 28), cmap='gray')
plt.show(block=True)
```

## Visualizing the output

```python
example_z = Variable(torch.FloatTensor(8*8*z_dim).normal_(0,1).view(64,-1))

samples = model.decoder(example_z).data.numpy()
fig = plt.figure(figsize=(4, 4))
gs = gridspec.GridSpec(8, 8)
gs.update(wspace=0.05, hspace=0.05)
for i, sample in enumerate(samples):
    ax = plt.subplot(gs[i])
    plt.axis('off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_aspect('equal')
    plt.imshow(sample.reshape(28, 28), cmap='Greys_r')
fig.savefig("generated_images_by_{}epochs.png".format(num_epoch))
```
